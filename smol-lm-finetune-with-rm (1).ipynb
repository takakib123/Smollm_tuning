```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from datasets import load_dataset
import torch
import os
```

```python
device =  "cuda"
```

```python
model_name = "HuggingFaceTB/SmolLM2-135M-Instruct"
model = AutoModelForCausalLM.from_pretrained(pretrained_model_name_or_path=model_name).to(device)
tokenizer = AutoTokenizer.from_pretrained(pretrained_model_name_or_path=model_name)
```

```python
messages = [{"role": "user", "content": "What is the capital of france"}]
input_text=tokenizer.apply_chat_template(messages, tokenize=False)
print(input_text)
inputs = tokenizer.encode(input_text, return_tensors="pt").to(device)
outputs = model.generate(inputs, max_new_tokens=50, temperature=0.2, top_p=0.9, do_sample=True)
print(outputs)
out = tokenizer.decode(outputs[0])
print(out)
```

```python
ds = load_dataset("HuggingFaceTB/smoltalk",'everyday-conversations')
```

```python
ds['train'][0]['messages']
```

```python
tokenizer.apply_chat_template(ds['train'][0]['messages'], tokenize = False)
```

```python
def tokenize_function(examples):
    texts = tokenizer.apply_chat_template(
            examples['messages'],
            tokenize=False
        )
    return tokenizer(texts, truncation=True, padding= False, max_length=512)
```

```python
tok_ds = ds.map(tokenize_function, batched=True)
```

```python
from transformers import DataCollatorForLanguageModeling
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm= False)
```

```python
tok_ds =tok_ds.remove_columns(["full_topic", "messages"])
```

```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(
    tok_ds["train"], shuffle=True, batch_size=8, collate_fn=data_collator
)
```

```python
from transformers import get_scheduler

num_epochs = 3
num_training_steps = num_epochs * len(train_dataloader)
lr_scheduler = get_scheduler(
    "linear",
    optimizer=optimizer,
    num_warmup_steps=0,
    num_training_steps=num_training_steps,
)
print(num_training_steps)
```

```python
model.train()
```

```python
def reward(r_embed, reward_model, out ):
    r_embed = r_tok.encode(out, return_tensors="pt")
    reward = reward_model(r_embed)
    reward = reward.logits.detach()
    return reward.squeeze(-1)
    
```

```python
from tqdm.auto import tqdm
import torch

# Assuming model, train_dataloader, optimizer, lr_scheduler, device, num_epochs, lamda are defined

num_training_steps = len(train_dataloader) * num_epochs
progress_bar = tqdm(range(num_training_steps))

model.train()
for epoch in range(num_epochs):
    for batch in train_dataloader:
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        output_text = model.generate(**batch)

        r = reward(r_embed,reward_model, tokenizer.decode(output_text[0]) ).to(0)

        loss = outputs.loss - r
        loss.backward()  
        print(loss)

        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()
        progress_bar.update(1)
```
